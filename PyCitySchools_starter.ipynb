```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_to_load = "C:/DataAnalytics/schools_complete.csv"
student_data_to_load = "C:/DataAnalytics/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## District Summary

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

```python

District_Summary= pd.DataFrame({"Total Schools" : school_data_complete['school_name'].nunique()},index=[0])
District_Summary["Total Students"]=school_data_complete["Student ID"].count()
District_Summary["Total Students"]=District_Summary["Total Students"].map("{:,}".format)
District_Summary["Total Budget"]=school_data["budget"].sum()
District_Summary['Total Budget'] = (District_Summary['Total Budget']).map("${:,.2f}".format)                                               
District_Summary["Average Math score"] = school_data_complete["math_score"].mean()
District_Summary["Average Reading score"] = school_data_complete["reading_score"].mean()
math_Passing_rate= (school_data_complete.loc[school_data_complete["math_score"] >= 70.0 ,:])
read_Passing_rate= (school_data_complete.loc[school_data_complete["reading_score"]>=70.0,:])
District_Summary["% Passing Math"] = (math_Passing_rate["Student ID"].count()/school_data_complete["Student ID"].count())*100
District_Summary["% Passing Reading"] = (read_Passing_rate["Student ID"].count()/school_data_complete["Student ID"].count())*100
District_Summary["% Overall Passing Rate"]= (District_Summary["Average Math score"]+ District_Summary["Average Reading score"]) /2
District_Summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
School_Based_report=school_data_complete.groupby(["school_name"])

#Get school wise Math Passing Rate 
math_score=school_data_complete.loc[school_data_complete["math_score"]>=70.0,:].groupby(["school_name"]).size(). rename('math_count').reset_index()
scp=school_data_complete.groupby(["school_name"]).size().rename('Total').reset_index()                              
read_score=school_data_complete.loc[school_data_complete["reading_score"]>=70.0,:].groupby(["school_name"]).size(). rename('reading_count').reset_index()


df = pd.merge(math_score, scp,how="inner" ,on=['school_name'])
df= pd.merge(df,read_score,how="inner",on =['school_name','school_name'])
df["% Passing Math"]=(df['math_count']/df['Total'])*100
df["% Passing Reading"]=(df['reading_count']/df['Total'])*100
df["% Overall Passing Rate"]= (df["% Passing Math"]+ df["% Passing Reading"])/2

# get school based report
School_Based_report_df=pd.DataFrame(School_Based_report["type"].max())

School_Based_report_df["Total Students"]=School_Based_report["Student ID"].count()
School_Based_report_df["Total School Budget"]=School_Based_report["budget"].mean().map("${:,.2f}".format)
School_Based_report_df["Per Student Budget"]=(School_Based_report["budget"].mean()/School_Based_report["Student ID"].count()).map("${:,.2f}".format)
School_Based_report_df["Average Math_score"]=School_Based_report["math_score"].mean()
School_Based_report_df["Average Reading score"]=School_Based_report["reading_score"].mean()

School_Based_report_df.rename(columns={"type":"School Type"},inplace=True)

# Perform Merge of both reports
school_summary=pd.merge(School_Based_report_df,df[['school_name','% Passing Math','% Passing Reading','% Overall Passing Rate']],how='inner',on=['school_name'])
school_summary=school_summary.sort_values(by="% Overall Passing Rate",ascending=False)
school_summary.reset_index(drop=True)
school_summary.set_index(['school_name'],inplace=True)
school_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_type_data=school_data_complete
school_type_data['math Pass']=[score >=70 for score in school_type_data['math_score']]
school_type_data['Reading Pass']=[score >=70 for score in school_type_data['reading_score']]

Data_by_size =school_type_data.groupby(["size"])
Data_by_size_df=pd.DataFrame(Data_by_size["math_score"].mean())

Data_by_size_df["Average reading score"]=Data_by_size["reading_score"].mean()
Data_by_size_df["% Passing Math"]=(Data_by_size["math Pass"].sum()/Data_by_size["Student ID"].count())*100
Data_by_size_df["%Passing Reading"]=(Data_by_size["Reading Pass"].sum()/Data_by_size["Student ID"].count())*100
Data_by_size_df["%Overall Passing rate"] =(Data_by_size_df["% Passing Math"]+ Data_by_size_df["%Passing Reading"])/2

Data_by_size_df=Data_by_size_df.reset_index()

Data_by_size_df["School Size"]=pd.cut(Data_by_size_df["size"],size_bins,labels=group_names)
Data_by_size_df=Data_by_size_df.drop(["size"], axis=1)
Data_by_size_df=Data_by_size_df.groupby(["School Size"])

Data_by_size_df.mean()
```
